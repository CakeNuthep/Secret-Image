```python
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from google.colab import files
```

```python
path_image = list(files.upload().keys())[0]
```

```python
image = cv.imread(path_image) 
cv2_imshow(image)
```

```python
lsb = int('0b00001111',2)
```

```python
secret_image = image & lsb
secret_image = secret_image << 4
cv2_imshow(secret_image)
```
